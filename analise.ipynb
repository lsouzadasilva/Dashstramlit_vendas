```python
import pandas as pd

DATA_URL = "https://raw.githubusercontent.com/lsouzadasilva/datasets/main/store_sales_2022-2023.csv"


df = pd.read_csv(DATA_URL)

display(df)

```

```python
df.info()
```

```python
df['date_of_sale'] = pd.to_datetime(df['date_of_sale'])
df.info()
```

```python
df['month'] = df['date_of_sale'].dt.month
df['year'] = df['date_of_sale'].dt.year
display(df)
```

## 5. Calculando Vendas por cidade

```python
YEAR = 2023

city_revenue = (
    df.groupby(['city', 'year'])['sales_amount']
    .sum()
    .unstack()
    .assign(change=lambda x: x.pct_change(axis=1)[YEAR] * 100) # percentual de diferenÃ§a comparado ao YEAR
)

city_revenue
```

## 6. Exemplo uma cidade

```python
city = "Tokyo"

revenue = city_revenue.loc[city, YEAR]
change = city_revenue.loc[city, "change"]

f"Revenue fot {city}: ${revenue:,.2f}, change: {change:.2f}%"
```

## 7. Preparando Dados

```python
select_city = "Tokyo"
select_year = 2023


filter_data = (
    df.query("city == @select_city and year == @select_year")
    .groupby("month", dropna=False, as_index=False)["sales_amount"]
    .sum()
)

filter_data
# type(filter_data)
```

```python
df["year_month"] = df["date_of_sale"].apply(lambda x: str(x.year) + "-" + str(x.month))
filtro = (df["product_name"] == "FlowFlare") | (df["city"] == "Osaka") | (df["year_month"] == "2023-3")
df = df[filtro]

total = df.groupby(["product_name","city","year_month"])[["quantity_sold"]].sum().reset_index()
display(total)
```
